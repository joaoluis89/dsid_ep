```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grpc_local = pd.read_csv("./Resultados/grpc/grpc_100exec_local.csv", index_col=0)
grpc_remoto = pd.read_csv("./Resultados/grpc/grpc_100exec_remoto.csv", index_col=0)
rpyc_local = pd.read_csv("./Resultados/rpyc/rpyc_100exec_local.csv", index_col=0)
rpyc_remoto = pd.read_csv("./Resultados/rpyc/rpyc_100exec_remoto.csv", index_col=0)
```

```python
grpc_local_trans = grpc_local.T
grpc_remoto_trans = grpc_remoto.T
rpyc_local_trans = rpyc_local.T
rpyc_remoto_trans = rpyc_remoto.T
```

```python
sns.boxplot(data=grpc_local_trans)
```

```python
sns.boxplot(data=grpc_remoto_trans)
```

```python
sns.boxplot(data=rpyc_local_trans)
```

```python
sns.boxplot(data=rpyc_remoto_trans)
```

```python
grpc_local_trans = grpc_local_trans.assign(MeioRpc="Local")
grpc_remoto_trans = grpc_remoto_trans.assign(MeioRpc="Remoto")
rpyc_local_trans = rpyc_local_trans.assign(MeioRpc="Local")
rpyc_remoto_trans = rpyc_remoto_trans.assign(MeioRpc="Remoto")
```

```python
grpc_local_trans = grpc_local_trans.drop(grpc_local_trans.voidCall.idxmax())
grpc_remoto_trans = grpc_remoto_trans.drop(grpc_remoto_trans.eightLongCall.idxmax())
```

```python
grpc_local_remoto = pd.concat([grpc_local_trans, grpc_remoto_trans])
grpc_local_remoto = pd.melt(grpc_local_remoto, id_vars="MeioRpc", value_vars=["voidCall", "longCall", "eightLongCall", "stringCall", "complexCall"])
```

```python
rpyc_local_remoto = pd.concat([rpyc_local_trans, rpyc_remoto_trans])
rpyc_local_remoto = pd.melt(rpyc_local_remoto, id_vars="MeioRpc", value_vars=["voidCall", "longCall", "eightLongCall", "stringCall", "complexCall"])
```
